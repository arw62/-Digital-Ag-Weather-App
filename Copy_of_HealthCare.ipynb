```python
# Install necessary libraries
!pip install matplotlib networkx

import matplotlib.pyplot as plt
import networkx as nx

# Step 1: Display Case Scenario
case_scenario = """
### Healthcare Root Cause Analysis (RCA) Simulation
*Scenario: Lab Test Delays Impacting Patient Care*

A hospital administrator receives complaints from physicians about delays in lab test results.
This is causing longer patient stays, treatment delays, and operational inefficiencies.

*Your task:*
1. Answer the "5 Whys" to determine the root cause.
2. Visualize your findings with a Fishbone Diagram.

"""

print(case_scenario)

# Step 2: Get Student Input for the 5 Whys
print("\n### Enter your '5 Whys' Analysis ###\n")

why1 = input("1️⃣ Why are lab test results delayed? → ")
why2 = input("2️⃣ Why is the lab processing tests slowly? → ")
why3 = input("3️⃣ Why is there a backlog of unprocessed tests? → ")
why4 = input("4️⃣ Why are there not enough lab technicians? → ")
why5 = input("5️⃣ Why were budget cuts made? → ")

whys = [why1, why2, why3, why4, why5]

# Display Student Responses
print("\n### Your 5 Whys Analysis ###")
for idx, why in enumerate(whys, start=1):
    print(f"{idx}. {why}")

# Step 3: Generate Fishbone Diagram from Student Input
def draw_fishbone(whys):
    G = nx.DiGraph()

    # Categories based on student responses
    categories = {
        "People": whys[3],
        "Process": "No priority system for urgent tests",
        "Technology": "Old lab equipment",
        "Environment": "High patient volume",
        "Policy": whys[4]
    }

    # Root cause
    root_cause = whys[0]

    # Add nodes
    G.add_node(root_cause, color='red')

    for key, value in categories.items():
        G.add_node(key, color='blue')
        G.add_edge(root_cause, key)

        # Sub-causes
        G.add_node(value, color='green')
        G.add_edge(key, value)

    # Draw the Fishbone Diagram
    plt.figure(figsize=(10, 6))
    pos = nx.spring_layout(G, seed=42)
    colors = [nx.get_node_attributes(G, 'color')[node] for node in G.nodes()]
    nx.draw(G, pos, with_labels=True, node_color=colors, edge_color='gray', node_size=3000, font_size=10, font_weight='bold')

    plt.title("Fishbone Diagram - Root Cause Analysis")
    plt.show()

print("\n### Generating Fishbone Diagram... ###")
draw_fishbone(whys)
```
